```python
import numpy as np
import scipy.signal as sgl
import pandas as pd
import matplotlib.pyplot as plt
from   scipy import special as spe
from matplotlib.figure import figaspect
from matplotlib.ticker import FormatStrFormatter
```

```python
#### exact phase velocity
def vphase_powerQ(f0, v0, alpha, Q0, df, fmax):
    """exact phase velocity from power-law Q""" 
    nhalf = int(fmax / df) + 1
    freq = np.arange(-nhalf,nhalf) * df
    
    omega  = 2 * np.pi * freq
    omega0 = 2 * np.pi * f0
    
    f_sign = np.sign(freq)
    
    Qw = Q0 * (np.abs(omega/omega0))**alpha * f_sign
    
    eps = np.finfo(float).eps
    
    ### avoid singularity for 1/Qw at Qw=0
    Qw[nhalf] = eps
    
    phi  = np.arctan(1/Qw)
    phi0 = np.arctan(1/Q0)

    ### hilbert transform
    analytic_signal = sgl.hilbert(phi)
    lnA = analytic_signal.imag
    
    ### lnA at freq = f0
    lnA0 = np.interp(f0, freq, lnA)
    
    diff_lnA = lnA - lnA0
    diff_phi = phi - phi0
    
    #### AdivA0 = A / A0 for freq >= 0
    AdivA0 = np.exp(diff_lnA - 1j*diff_phi )[nhalf:]

    M0 = v0**2 * (1 - 1j/Q0)
    
    M = M0 * AdivA0
    
    vc = np.sqrt(M)
    
    vr = vc.real
    vi = vc.imag
    
    vphase = (vr**2 + vi**2) / vr
    
    return freq[nhalf:], vphase
```

```python
###### GSLS weight function

def WR(tauEpsilon, tauSigma, omega):
    """The dispersive weight function  """   
    if np.isscalar(omega):  ### It means that omega is a single number        
        tmp1 = omega*tauEpsilon
        tmp2 = omega*tauSigma
        
        res = (1 + tmp1*tmp2) / (1 + tmp2**2)
        
        WR_omega = np.sum(res)
    
    else: ### It means that omega is an array
        ### tmp1 and tmp2 shape: (omega.shape, tauEps.shape)
        tmp1 = np.multiply.outer(omega, tauEpsilon)    
        tmp2 = np.multiply.outer(omega, tauSigma)

        ### the fraction function for each omega and each 
        res = (1 + tmp1*tmp2) / (1 + tmp2**2)  
        
        WR_omega = np.sum(res, axis=1)            
    
    return WR_omega   
    
    
def WI(tauEpsilon, tauSigma, omega):
    """The dispersive weight function  """   
    if np.isscalar(omega):  ### It means that omega is a single number        
        tmp1 = omega*(tauEpsilon-tauSigma)
        tmp2 = omega*tauSigma
        
        res = tmp1 / (1 + tmp2**2)
        
        WI_omega = np.sum(res)
    
    else: ### It means that omega is an array
        ### tmp1 and tmp2 shape: (omega.shape, tauEps.shape)
        tmp1 = np.multiply.outer(omega, tauEpsilon-tauSigma)    
        tmp2 = np.multiply.outer(omega, tauSigma)

        ### the fraction function for each omega and each 
        res = tmp1 / (1 + tmp2**2)  
        
        WI_omega = np.sum(res, axis=1)            
    
    return WI_omega   


def cmodulus_Kjart(M0, iQ, omega0, omega):
    """ Complex modulus for the Kjartansson model """
    img = 1j
    
    gamma = np.arctan(iQ) / np.pi
    
#     m = M0 / np.cos(np.pi * gamma)
    
    modulus = M0 * np.power(img*omega/omega0, 2.*gamma)
    
    return modulus


def cmodulus_Kolsky(M0, iQ, omega0, omega):
    """ Complex modulus for the Futermann model """
    img = 1j
    
    modulus = M0*(1 + (2/np.pi*np.log(omega/omega0) + img)*iQ)
    
    return modulus
```

```python
### power law Q models after calibration
### M0*(1-iQ0) denotes the complex modulus at omega0
### ------------------------------------------------
def Mpowerlaw_2nd(M0, iQ0, alpha, omega0, omega): 
    ### Mij for the second-order nearly constant Q model
    img = 1j
    
    PI = np.pi
     
    den   = (4.+(2.*(iQ0**2)))-(alpha*(iQ0*PI))
    a1num = 4
    a2num = (2.*iQ0)-(alpha*PI)
    
    a1 = a1num / den
    a2 = a2num / den
                  
    WC = (2./PI)*np.log(omega/omega0) - img
    
    M = 1 + iQ0*(a1*WC + a2*WC**2)
    
    ### calibration
    Mref = M0 / (1 + iQ0*(-a2))
    
    M *= Mref
    
    return M


def Mpowerlaw_2nd_appr(M0, iQ0, alpha, tauSigma, tauEpsilon, omega0, omega): 
    ### Mij for the second-order nearly constant Q model
    img = 1j
    
    Wreal  = WR(tauEpsilon, tauSigma, omega)
    Wreal0 = WR(tauEpsilon, tauSigma, omega0)
    
    Wimag  = WI(tauEpsilon, tauSigma, omega)
    
    WC     = Wreal - Wreal0 - img*Wimag
    
    PI = np.pi
     
    den   = (4.+(2.*(iQ0**2)))-(alpha*(iQ0*PI))
    a1num = 4
    a2num = (2.*iQ0)-(alpha*PI)
    
    a1 = a1num / den
    a2 = a2num / den
                  
    M = 1 + iQ0*(a1*WC + a2*WC**2)
    
    ## calibration
    WC0I = WI(tauEpsilon, tauSigma, omega0)
    Mref  = M0 / (1 + iQ0*(-a2*WC0I**2)) 
#     Mref = M0 / (1 + iQ0*(-a2))
    
    M *= Mref    
    
    return M


def Mpowerlaw_3rd(M0, iQ0, alpha, omega0, omega): 
    ### Mij for the second-order nearly constant Q model
    img = 1j
    
    PI = np.pi
     
    den   = 24.+((-14.*(alpha*(iQ0*PI)))+((iQ0**2)*(20.+((alpha**2)*(PI**2)))))
    a1num = 24.+((12.*(iQ0**2))+((-14.*(alpha*(iQ0*PI)))+((alpha**2)*(PI**2))))
    a2num = (-6.*(alpha*PI))+(iQ0*(12.+((alpha**2)*(PI**2))))
    a3num = (4.*(iQ0**2))+((-6.*(alpha*(iQ0*PI)))+((alpha**2)*(PI**2)))
    
    a1 = a1num / den
    a2 = a2num / den
    a3 = a3num / den
                  
    WC = (2./PI)*np.log(omega/omega0) - img
    
    M = 1 + iQ0*(a1*WC + a2*WC**2 + a3*WC**3)
    
    ### calibration
    Mref = M0 / (1 + iQ0*(-a2)) 
    
    M *= Mref
    
    return M


def Mpowerlaw_3rd_appr(Mref, iQ0, alpha, tauSigma, tauEpsilon, omega0, omega): 
    ### Mij for the second-order nearly constant Q model
    img = 1j
    
    Wreal  = WR(tauEpsilon, tauSigma, omega)
    Wreal0 = WR(tauEpsilon, tauSigma, omega0)
    
    Wimag  = WI(tauEpsilon, tauSigma, omega)
    
    WC     = Wreal - Wreal0 - img*Wimag
    
    PI = np.pi
     
    den   = 24.+((-14.*(alpha*(iQ0*PI)))+((iQ0**2)*(20.+((alpha**2)*(PI**2)))))
    a1num = 24.+((12.*(iQ0**2))+((-14.*(alpha*(iQ0*PI)))+((alpha**2)*(PI**2))))
    a2num = (-6.*(alpha*PI))+(iQ0*(12.+((alpha**2)*(PI**2))))
    a3num = (4.*(iQ0**2))+((-6.*(alpha*(iQ0*PI)))+((alpha**2)*(PI**2)))
    
    a1 = a1num / den
    a2 = a2num / den
    a3 = a3num / den
                  
    M = 1 + iQ0*(a1*WC + a2*WC**2 + a3*WC**3)
    
    ### calibration
    WC0I = WI(tauEpsilon, tauSigma, omega0)
    Mref  = M0 / (1 + iQ0*(-a2*WC0I**2))
#     Mref = M0 / (1 + iQ0*(-a2))
    
    M *= Mref    
    
    return M


def Mpowerlaw_4th(M0, iQ0, alpha, omega0, omega): 
    ### Mij for the second-order nearly constant Q model
    img = 1j
    
    PI = np.pi
     
    aux0 = (372.*((alpha**2)*(iQ0*(PI**2))))+(-2.*((alpha**3.)*((3.+(5.*(iQ0**2)))*(PI**3.))));
    aux1 = (48.*(iQ0*(28.+(5.*(iQ0**2)))))+((-48.*(alpha*((26.+(11.*(iQ0**2)))*PI)))+aux0);
    den  = 1152.+(iQ0*(aux1-((alpha**4.)*(iQ0*(PI**4.)))));
    
    aux0  = (60.*(iQ0**2))+((-78.*(alpha*(iQ0*PI)))+(3.*((alpha**2)*((1.+(3.*(iQ0**2)))*(PI**2)))));
    a1num = 16.*((72.+aux0)-((alpha**3.)*(iQ0*(PI**3.))));
    
    aux0  = (-12.*((alpha**3.)*((1.+(iQ0**2))*(PI**3.))))+(-2.*((alpha**4.)*(iQ0*(PI**4.))));
    aux1  = (-96.*(alpha*((3.+(7.*(iQ0**2)))*PI)))+((360.*((alpha**2)*(iQ0*(PI**2))))+aux0);
    a2num = (288.*(iQ0*(2.+(iQ0**2))))+aux1;
    
    aux0  = (12.*(iQ0**2))+((-18.*(alpha*(iQ0*PI)))+(3.*((alpha**2)*((1.+(iQ0**2))*(PI**2)))));
    a3num = 16.*(aux0-((alpha**3.)*(iQ0*(PI**3.))));
    
    aux0  = (84.*((alpha**2)*(iQ0*(PI**2))))+(-2.*((alpha**3.)*((3.+(iQ0**2))*(PI**3.))));
    a4num = ((48.*(iQ0**3.))+((-144.*(alpha*((iQ0**2)*PI)))+aux0))-((alpha**4.)*(iQ0*(PI**4.)));
    
    a1 = a1num / den
    a2 = a2num / den
    a3 = a3num / den
    a4 = a4num / den
                  
    WC = (2./PI)*np.log(omega/omega0) - img
    
    M = 1 + iQ0*(a1*WC + a2*WC**2 + a3*WC**3 + a4*WC**4)
    
    ### calibration
    Mref = M0 / (1 + iQ0*(-a2 + a4))     
    
    M *= Mref
    
    return M

def Mpowerlaw_4th_appr(Mref, iQ0, alpha, tauSigma, tauEpsilon, omega0, omega): 
    ### Mij for the second-order nearly constant Q model
    img = 1j
    
    Wreal  = WR(tauEpsilon, tauSigma, omega)
    Wreal0 = WR(tauEpsilon, tauSigma, omega0)
    
    Wimag  = WI(tauEpsilon, tauSigma, omega)
    
    WC     = Wreal - Wreal0 - img*Wimag
    
    PI = np.pi
     
    aux0 = (372.*((alpha**2)*(iQ0*(PI**2))))+(-2.*((alpha**3.)*((3.+(5.*(iQ0**2)))*(PI**3.))));
    aux1 = (48.*(iQ0*(28.+(5.*(iQ0**2)))))+((-48.*(alpha*((26.+(11.*(iQ0**2)))*PI)))+aux0);
    den  = 1152.+(iQ0*(aux1-((alpha**4.)*(iQ0*(PI**4.)))));
    
    aux0  = (60.*(iQ0**2))+((-78.*(alpha*(iQ0*PI)))+(3.*((alpha**2)*((1.+(3.*(iQ0**2)))*(PI**2)))));
    a1num = 16.*((72.+aux0)-((alpha**3.)*(iQ0*(PI**3.))));
    
    aux0  = (-12.*((alpha**3.)*((1.+(iQ0**2))*(PI**3.))))+(-2.*((alpha**4.)*(iQ0*(PI**4.))));
    aux1  = (-96.*(alpha*((3.+(7.*(iQ0**2)))*PI)))+((360.*((alpha**2)*(iQ0*(PI**2))))+aux0);
    a2num = (288.*(iQ0*(2.+(iQ0**2))))+aux1;
    
    aux0  = (12.*(iQ0**2))+((-18.*(alpha*(iQ0*PI)))+(3.*((alpha**2)*((1.+(iQ0**2))*(PI**2)))));
    a3num = 16.*(aux0-((alpha**3.)*(iQ0*(PI**3.))));
    
    aux0  = (84.*((alpha**2)*(iQ0*(PI**2))))+(-2.*((alpha**3.)*((3.+(iQ0**2))*(PI**3.))));
    a4num = ((48.*(iQ0**3.))+((-144.*(alpha*((iQ0**2)*PI)))+aux0))-((alpha**4.)*(iQ0*(PI**4.)));
    
    a1 = a1num / den
    a2 = a2num / den
    a3 = a3num / den
    a4 = a4num / den
                  
    M = 1 + iQ0*(a1*WC + a2*WC**2 + a3*WC**3 + a4*WC**4)
    
    ### calibration
    WC0I = WI(tauEpsilon, tauSigma, omega0)
    Mref  = M0 / (1 + iQ0*(-a2*WC0I**2 + a4*WC0I**4)) 
#     Mref = M0 / (1 + iQ0*(-a2 + a4))
    
    M *= Mref    
    
    return M


def Mpowerlaw_5th(M0, iQ0, alpha, omega0, omega): 
    ### Mij for the second-order nearly constant Q model
    img = 1j
    
    PI = np.pi

    aux0 = (24.*((alpha**4.)*(iQ0*((3.+(5.*(iQ0**2)))*(PI**4.)))))+((54.*((alpha**5.)*((iQ0**2)*(PI**5.))))+((alpha**6.)*(iQ0*(PI**6.))));
    aux1 = (144.*((alpha**2)*(iQ0*((467.+(58.*(iQ0**2)))*(PI**2)))))+((-24.*((alpha**3.)*((47.+(383.*(iQ0**2)))*(PI**3.))))+aux0);
    aux2 = (576.*(iQ0*(180.+(61.*(iQ0**2)))))+((-576.*(alpha*((210.+(157.*(iQ0**2)))*PI)))+aux1);
    den  = 69120.+(iQ0*aux2);
        
    aux0  = (-12.*((alpha**4.)*((-7.+(iQ0**2))*(PI**4.))))+((54.*((alpha**5.)*(iQ0*(PI**5.))))+((alpha**6.)*((iQ0**2)*(PI**6.))));
    aux1  = (144.*((alpha**2)*((20.+(437.*(iQ0**2)))*(PI**2))))+((-24.*((alpha**3.)*(iQ0*((257.+(101.*(iQ0**2)))*(PI**3.)))))+aux0);
    a1num =(2880.*(24.+((28.*(iQ0**2))+(5.*(iQ0**4.)))))+((-1728.*(alpha*(iQ0*((70.+(29.*(iQ0**2)))*PI))))+aux1);    
    
    aux0  = (84.*((alpha**5.)*((iQ0**2)*(PI**5.))))+(2.*((alpha**6.)*(iQ0*(PI**6.))));
    aux1  = (-48.*((alpha**3.)*((15.+(197.*(iQ0**2)))*(PI**3.))))+((48.*((alpha**4.)*(iQ0*((8.+(3.*(iQ0**2)))*(PI**4.)))))+aux0);
    aux2  = (-17280.*(alpha*((1.+(4.*(iQ0**2)))*PI)))+((288.*((alpha**2)*(iQ0*((115.+(34.*(iQ0**2)))*(PI**2)))))+aux1);
    a2num =(5760.*(iQ0*(6.+(5.*(iQ0**2)))))+aux2;
    
    aux0  = (12.*((alpha**4.)*((5.+(7.*(iQ0**2)))*(PI**4.))))+((42.*((alpha**5.)*(iQ0*(PI**5.))))+((alpha**6.)*((iQ0**2)*(PI**6.))));
    aux1  = (720.*((alpha**2)*((2.+(19.*(iQ0**2)))*(PI**2))))+((-72.*((alpha**3.)*(iQ0*((45.+(7.*(iQ0**2)))*(PI**3.)))))+aux0);
    aux2  = (2880.*((iQ0**2)*(2.+(iQ0**2))))+((-2880.*(alpha*(iQ0*((3.+(4.*(iQ0**2)))*PI))))+aux1);
    a3num = 2.*aux2;
    
    aux0  = (24.*((alpha**4.)*(iQ0*((5.+(iQ0**2))*(PI**4.)))))+((30.*((alpha**5.)*((iQ0**2)*(PI**5.))))+((alpha**6.)*(iQ0*(PI**6.))));
    aux1  = (720.*((alpha**2)*(iQ0*((7.+(2.*(iQ0**2)))*(PI**2)))))+((-360.*((alpha**3.)*((1.+(5.*(iQ0**2)))*(PI**3.))))+aux0);
    a4num = (2880.*(iQ0**3.))+((-8640.*(alpha*((iQ0**2)*PI)))+aux1);
    
    aux0  = (-12.*((alpha**4.)*((-3.+(iQ0**2))*(PI**4.))))+((30.*((alpha**5.)*(iQ0*(PI**5.))))+((alpha**6.)*((iQ0**2)*(PI**6.))));
    aux1  = (3600.*((alpha**2)*((iQ0**2)*(PI**2))))+((-120.*((alpha**3.)*(iQ0*((9.+(iQ0**2))*(PI**3.)))))+aux0);
    a5num = (576.*(iQ0**4.))+((-2880.*(alpha*((iQ0**3.)*PI)))+aux1);
    
    a1 = a1num / den
    a2 = a2num / den
    a3 = a3num / den
    a4 = a4num / den
    a5 = a5num / den
                  
    WC = (2./PI)*np.log(omega/omega0) - img
    
    M = 1 + iQ0*(a1*WC + a2*WC**2 + a3*WC**3 + a4*WC**4 + a5*WC**5)
    
    ### calibration
    Mref = M0 / (1 + iQ0*(-a2 + a4))      
    
    M *= Mref
    
    return M


def Mpowerlaw_5th_appr(Mref, iQ0, alpha, tauSigma, tauEpsilon, omega0, omega): 
    ### Mij for the second-order nearly constant Q model
    img = 1j
    
    Wreal  = WR(tauEpsilon, tauSigma, omega)
    Wreal0 = WR(tauEpsilon, tauSigma, omega0)
    
    Wimag  = WI(tauEpsilon, tauSigma, omega)
    
    WC     = Wreal - Wreal0 - img*Wimag
    
    PI = np.pi
     
    aux0 = (24.*((alpha**4.)*(iQ0*((3.+(5.*(iQ0**2)))*(PI**4.)))))+((54.*((alpha**5.)*((iQ0**2)*(PI**5.))))+((alpha**6.)*(iQ0*(PI**6.))));
    aux1 = (144.*((alpha**2)*(iQ0*((467.+(58.*(iQ0**2)))*(PI**2)))))+((-24.*((alpha**3.)*((47.+(383.*(iQ0**2)))*(PI**3.))))+aux0);
    aux2 = (576.*(iQ0*(180.+(61.*(iQ0**2)))))+((-576.*(alpha*((210.+(157.*(iQ0**2)))*PI)))+aux1);
    den  = 69120.+(iQ0*aux2);
        
    aux0  = (-12.*((alpha**4.)*((-7.+(iQ0**2))*(PI**4.))))+((54.*((alpha**5.)*(iQ0*(PI**5.))))+((alpha**6.)*((iQ0**2)*(PI**6.))));
    aux1  = (144.*((alpha**2)*((20.+(437.*(iQ0**2)))*(PI**2))))+((-24.*((alpha**3.)*(iQ0*((257.+(101.*(iQ0**2)))*(PI**3.)))))+aux0);
    a1num =(2880.*(24.+((28.*(iQ0**2))+(5.*(iQ0**4.)))))+((-1728.*(alpha*(iQ0*((70.+(29.*(iQ0**2)))*PI))))+aux1);    
    
    aux0  = (84.*((alpha**5.)*((iQ0**2)*(PI**5.))))+(2.*((alpha**6.)*(iQ0*(PI**6.))));
    aux1  = (-48.*((alpha**3.)*((15.+(197.*(iQ0**2)))*(PI**3.))))+((48.*((alpha**4.)*(iQ0*((8.+(3.*(iQ0**2)))*(PI**4.)))))+aux0);
    aux2  = (-17280.*(alpha*((1.+(4.*(iQ0**2)))*PI)))+((288.*((alpha**2)*(iQ0*((115.+(34.*(iQ0**2)))*(PI**2)))))+aux1);
    a2num =(5760.*(iQ0*(6.+(5.*(iQ0**2)))))+aux2;
    
    aux0  = (12.*((alpha**4.)*((5.+(7.*(iQ0**2)))*(PI**4.))))+((42.*((alpha**5.)*(iQ0*(PI**5.))))+((alpha**6.)*((iQ0**2)*(PI**6.))));
    aux1  = (720.*((alpha**2)*((2.+(19.*(iQ0**2)))*(PI**2))))+((-72.*((alpha**3.)*(iQ0*((45.+(7.*(iQ0**2)))*(PI**3.)))))+aux0);
    aux2  = (2880.*((iQ0**2)*(2.+(iQ0**2))))+((-2880.*(alpha*(iQ0*((3.+(4.*(iQ0**2)))*PI))))+aux1);
    a3num = 2.*aux2;
    
    aux0  = (24.*((alpha**4.)*(iQ0*((5.+(iQ0**2))*(PI**4.)))))+((30.*((alpha**5.)*((iQ0**2)*(PI**5.))))+((alpha**6.)*(iQ0*(PI**6.))));
    aux1  = (720.*((alpha**2)*(iQ0*((7.+(2.*(iQ0**2)))*(PI**2)))))+((-360.*((alpha**3.)*((1.+(5.*(iQ0**2)))*(PI**3.))))+aux0);
    a4num = (2880.*(iQ0**3.))+((-8640.*(alpha*((iQ0**2)*PI)))+aux1);
    
    aux0  = (-12.*((alpha**4.)*((-3.+(iQ0**2))*(PI**4.))))+((30.*((alpha**5.)*(iQ0*(PI**5.))))+((alpha**6.)*((iQ0**2)*(PI**6.))));
    aux1  = (3600.*((alpha**2)*((iQ0**2)*(PI**2))))+((-120.*((alpha**3.)*(iQ0*((9.+(iQ0**2))*(PI**3.)))))+aux0);
    a5num = (576.*(iQ0**4.))+((-2880.*(alpha*((iQ0**3.)*PI)))+aux1);
    
    a1 = a1num / den
    a2 = a2num / den
    a3 = a3num / den
    a4 = a4num / den
    a5 = a5num / den
    
    M = 1 + iQ0*(a1*WC + a2*WC**2 + a3*WC**3 + a4*WC**4 + a5*WC**5)
    
    ### calibration
    WC0I = WI(tauEpsilon, tauSigma, omega0)
    Mref  = M0 / (1 + iQ0*(-a2*WC0I**2 + a4*WC0I**4)) 
#     Mref = M0 / (1 + iQ0*(-a2 + a4))  
    
    M *= Mref
    
    return M
```

## Optimial parameters for the weighting function in frequency range [1, 200] Hz

```python
#### tauSigma and tauEpsilon for the frequency range [1, 200] Hz
tauSigma1to200   = np.array([1.8230838e-01, 3.2947348e-02, 8.4325390e-03, 2.3560480e-03, 5.1033826e-04])
dtau1to200       = np.array([2.7518001e-01, 3.0329269e-02, 6.9820198e-03, 1.9223614e-03, 7.2390630e-04])
tauEpsilon1to200 = tauSigma1to200 + dtau1to200
```

## Power law Q in the frequency range [0.01, 2] Hz

```python
### min and max frequencies 
fmin = 0.01
fmax = 3

#### Q0 is taken as 50, 100, 300
Q0 = 100

### dominant frequency
f0     = 1

freq   = np.linspace(fmin, fmax, num=101)

### scale the [1, 200]-Hz coefficients to those for [0.015, 3] Hz.
scale_fac  = 1/50
tauSigma   = tauSigma1to200 / scale_fac
tauEpsilon = tauEpsilon1to200 / scale_fac

omega0 = 2 * np.pi * f0
omega  = 2 * np.pi * freq

iQ0    = 1/Q0

#### reference velocity
v0 = 3
M0 = v0**2

### alpha=0.2, 0.5, 0.8
alpha  = 0.2

###imaginary unit
img = 1j

### exact power-law Q 
Qexact = Q0 * (omega/omega0)**alpha

### exact phase velocity for power-law Q
fmax_exact = 6
df_exact   = 10**-5
fexact, Vexact = vphase_powerQ(f0, v0, alpha, Q0, df_exact, fmax_exact)


### 2nd order model
Mc2nd = Mpowerlaw_2nd(M0, iQ0, alpha, omega0, omega)
Q2nd  = - Mc2nd.real / Mc2nd.imag

vc    = np.sqrt(Mc2nd)    
V2nd  = (vc.real**2 + vc.imag**2) / vc.real

### 2nd order model approximation
Mc2nd_appr = Mpowerlaw_2nd_appr(M0, iQ0, alpha, tauSigma, tauEpsilon, omega0, omega)
Q2nd_appr  = - Mc2nd_appr.real / Mc2nd_appr.imag

vc         = np.sqrt(Mc2nd_appr) 
V2nd_appr  = (vc.real**2 + vc.imag**2) / vc.real

### 3rd order model
Mc3rd = Mpowerlaw_3rd(M0, iQ0, alpha, omega0, omega)
Q3rd  = - Mc3rd.real / Mc3rd.imag

vc    = np.sqrt(Mc3rd)    
V3rd  = (vc.real**2 + vc.imag**2) / vc.real

### 3rd order model approximation
Mc3rd_appr = Mpowerlaw_3rd_appr(M0, iQ0, alpha, tauSigma, tauEpsilon, omega0, omega)
Q3rd_appr  = - Mc3rd_appr.real / Mc3rd_appr.imag

vc         = np.sqrt(Mc3rd_appr)  
V3rd_appr  = (vc.real**2 + vc.imag**2) / vc.real


### 4th order model
Mc4th = Mpowerlaw_4th(M0, iQ0, alpha, omega0, omega)
Q4th  = - Mc4th.real / Mc4th.imag

vc    = np.sqrt(Mc4th) 
V4th  = (vc.real**2 + vc.imag**2) / vc.real


### 4th order model approximation
Mc4th_appr = Mpowerlaw_4th_appr(M0, iQ0, alpha, tauSigma, tauEpsilon, omega0, omega)
Q4th_appr  = - Mc4th_appr.real / Mc4th_appr.imag

vc         = np.sqrt(Mc4th_appr) 
V4th_appr  = (vc.real**2 + vc.imag**2) / vc.real

### 5th order model
Mc5th = Mpowerlaw_5th(M0, iQ0, alpha, omega0, omega)
Q5th  = - Mc5th.real / Mc5th.imag

vc    = np.sqrt(Mc5th) 
V5th  = (vc.real**2 + vc.imag**2) / vc.real

### 5th order model approximation
Mc5th_appr = Mpowerlaw_5th_appr(M0, iQ0, alpha, tauSigma, tauEpsilon, omega0, omega)
Q5th_appr  = - Mc5th_appr.real / Mc5th_appr.imag

vc         = np.sqrt(Mc5th_appr) 
V5th_appr  = (vc.real**2 + vc.imag**2) / vc.real
```

## Plot Q curves for 2nd order model

```python
#####Plot Q curves ###################################################
fig = plt.figure(figsize=(4, 2.5))
ax = fig.add_subplot(1, 1, 1)

#fig.subplots_adjust(left=.18, bottom=.16, right=.95, top=.97)

plt.plot(freq, Qexact, color="gray", lw=1, ls="solid",  \
         label="exact")
plt.plot(freq, Q2nd, color="red", lw=1, ls="dotted",  \
         label="2nd")
plt.plot(freq, Q2nd_appr, color="lime", lw=1, ls="dotted",  \
         label="2nd ana")

plt.xlabel("f (Hz)", fontsize=16)
plt.ylabel("Q", fontsize=16)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

ax.legend(loc="upper left", ncol=1, fontsize=8, frameon=False)

ax.set_xlim((fmin,fmax))

if alpha >= 0.6:
    ymin = -10
    ymax = 273
elif 0.3 < alpha < 0.6:
    ymin = 1.8
    ymax = 181.4
else:
    ymin = 35.5
    ymax = 129.3   

ax.set_ylim((ymin,ymax))

### add the vertical line and text for the reference frequency
plt.axvline(x=f0, label='$f_{0}$ ='.format(f0), lw=0.5, color="magenta", ls=(0,(5,10)))
labeltext = r"$f_{0}$"
limy = ax.get_ylim()
ymin = limy[0]
ymax = limy[1]
ypos = ymin + (ymax-ymin)/30
plt.text(f0, ypos, labeltext) 

fnQ    = "./figs/Q2nd_Q%s_alpha%s_fc%s.pdf" %(str(Q0), str(alpha), str(f0))
fig.savefig(fnQ, bbox_inches="tight")
       
plt.show()
plt.close()

####################################
fig = plt.figure(figsize=(4, 2.5))
ax = fig.add_subplot(1, 1, 1)

#fig.subplots_adjust(left=.18, bottom=.16, right=.95, top=.97)

plt.plot(fexact, Vexact, color="gray", lw=1, ls="solid",  \
         label="exact")
plt.plot(freq, V2nd, color="red", lw=1, ls="dotted",  \
         label="2nd")
plt.plot(freq, V2nd_appr, color="lime", lw=1, ls="dotted",  \
         label="2nd ana")

plt.xlabel("f (Hz)", fontsize=16)
plt.ylabel("V (km/s)", fontsize=16)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

ax.legend(loc="upper left", ncol=1, fontsize=8, frameon=False)

# ax.set_xscale('log')

if alpha >= 1:
    ax.legend(loc="upper left", ncol=1, fontsize=8, frameon=False)
elif 0.5 <= alpha < 1:
    ax.legend(loc="upper left", ncol=1, fontsize=8, frameon=False)
else:
    ax.legend(loc="upper left", ncol=1, fontsize=8, frameon=False)

if alpha >= 1:
    ymin = 2.95
    ymax = 3.05
elif 0.5 < alpha < 1:
    ymin = 2.95
    ymax = 3.05
else:
    ymin = 2.95
    ymax = 3.05
    
ax.set_xlim((fmin,fmax))
ax.set_ylim((ymin,ymax))

### add the vertical line and text for the reference frequency
plt.axvline(x=f0, label='$f_{0}$ ='.format(f0), lw=0.5, color="magenta", ls=(0,(5,10)))
labeltext = r"$f_{0}$"
ypos = ymin + (ymax-ymin)/30
plt.text(f0, ypos, labeltext) 

fnQ    = "./figs/V2nd_Q%s_alpha%s_fc%s.pdf" %(str(Q0), str(alpha), str(f0))
fig.savefig(fnQ, bbox_inches="tight")
       
plt.show()
plt.close()
```
