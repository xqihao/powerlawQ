```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
###### GSLS weight function

def WR(tauEpsilon, tauSigma, omega):
    """The dispersive weight function  """   
    if np.isscalar(omega):  ### It means that omega is a single number        
        tmp1 = omega*tauEpsilon
        tmp2 = omega*tauSigma
        
        res = (1 + tmp1*tmp2) / (1 + tmp2**2)
        
        WR_omega = np.sum(res)
    
    else: ### It means that omega is an array
        ### tmp1 and tmp2 shape: (omega.shape, tauEps.shape)
        tmp1 = np.multiply.outer(omega, tauEpsilon)    
        tmp2 = np.multiply.outer(omega, tauSigma)

        ### the fraction function for each omega and each 
        res = (1 + tmp1*tmp2) / (1 + tmp2**2)  
        
        WR_omega = np.sum(res, axis=1)            
    
    return WR_omega   
    
    
def WI(tauEpsilon, tauSigma, omega):
    """The dispersive weight function  """   
    if np.isscalar(omega):  ### It means that omega is a single number        
        tmp1 = omega*(tauEpsilon-tauSigma)
        tmp2 = omega*tauSigma
        
        res = tmp1 / (1 + tmp2**2)
        
        WI_omega = np.sum(res)
    
    else: ### It means that omega is an array
        ### tmp1 and tmp2 shape: (omega.shape, tauEps.shape)
        tmp1 = np.multiply.outer(omega, tauEpsilon-tauSigma)    
        tmp2 = np.multiply.outer(omega, tauSigma)

        ### the fraction function for each omega and each 
        res = tmp1 / (1 + tmp2**2)  
        
        WI_omega = np.sum(res, axis=1)            
    
    return WI_omega  

var_dict = {}

def cmodulus_1st2nd_model_init(tauEpsilon, tauSigma):
    ### initialize the weight inversion function
        
    var_dict["tauEpsilon"] = tauEpsilon

    var_dict["tauSigma"]   = tauSigma      
    
    return

def cmodulus_1stmodel(M0, iQ0, omega0, omega):
    """ Complex modulus for the 1st-order model """
    img = 1j
    
    tauEpsilon = var_dict["tauEpsilon"]
    tauSigma   = var_dict["tauSigma"]    
    
    Wc = WR(tauEpsilon, tauSigma, omega) - WR(tauEpsilon, tauSigma, omega0) \
     + WI(tauEpsilon, tauSigma, omega)*(-img)
    
    M_1st = M0 * (1 + iQ0*Wc)
    
    return M_1st
    

def cmodulus_2ndmodel(M0, iQ0, omega0, omega):
    """ Complex modulus for the 2nd-order model """
    img = 1j
    
    tauEpsilon = var_dict["tauEpsilon"]
    tauSigma   = var_dict["tauSigma"]
    
    Wc = WR(tauEpsilon, tauSigma, omega) - WR(tauEpsilon, tauSigma, omega0) \
     + WI(tauEpsilon, tauSigma, omega)*(-img)
    
    M_2nd = M0 * (1 + iQ0*Wc + 1/2*(iQ0*Wc)**2)
    
    return M_2nd

def cmodulus_2ndmodelcalb(M0, iQ0, omega0, omega):
    """ Complex modulus for the 2nd-order model calibrated """
    img = 1j
    
    tauEpsilon = var_dict["tauEpsilon"]
    tauSigma   = var_dict["tauSigma"]
    
    ### M and iQ at omega=omega0 in the Kjartansson model
    Mr_kjar = np.real(cmodulus_Kjart(M0, iQ0, omega0, omega0))
    iQ_kjar = iQ0
    
    ### calbirated M0 and iQ0
    iQ0_calb = 2*iQ_kjar / (1 + np.sqrt(1+2*iQ_kjar**2))    
    M0_calb  = Mr_kjar / (1-0.5*iQ0_calb**2)
    
    Wc = WR(tauEpsilon, tauSigma, omega) - WR(tauEpsilon, tauSigma, omega0) \
     + WI(tauEpsilon, tauSigma, omega)*(-img)
    
    M_2nd = M0_calb * (1 + iQ0_calb*Wc + 1/2*(iQ0_calb*Wc)**2)
    
    return M_2nd


def cmodulus_Kjart(M0, iQ0, omega0, omega):
    """ Complex modulus for the Kjartansson model """
    img = 1j
    
    gamma = np.arctan(iQ0) / np.pi
    
    modulus = M0 * np.power(-img*omega/omega0, 2.*gamma)
    
    return modulus


def cmodulus_Kolsk(M0, iQ0, omega0, omega):
    """ Complex modulus for the Futermann model """
    img = 1j
    
    if np.isscalar(omega):  ###scalar 
        if omega == 0.:
            modulus = M0
    else: ###array
        ####  find the index of the omega == 0 
        index_arr = np.where(omega==0)[0]   
        index0 = index_arr[0]

        if index0.size != 0: ## it means that 0 exists
            omega_new = omega.copy()
            omega_new[index0] = 1.0E-6 
            
            modulus = M0*(1 + (2/np.pi*np.log(np.abs(omega_new/omega0)) - img*np.sign(omega))*iQ0)
            modulus[index0] = M0
        else:
            modulus = M0*(1 + (2/np.pi*np.log(np.abs(omega/omega0)) - img*np.sign(omega))*iQ0)
    
    
    return modulus


def point_source_wave_atten(modufun, M0, iQ0, omega0, omega, r):
    """ green function for the attenuating models """
    img = 1j
    
    M = modufun(M0, iQ0, omega0, omega)
    
    vel = np.sqrt(M)
    
    ####  find the index of the omega == 0 
    index_arr = np.where(omega==0)[0]   
    index0 = index_arr[0]
    
    ###  set a nonzero value to avoid the error caused by the division by zero.       
    M[index0] = 1.
    vel[index0] = 1.    
       
    P = np.exp(img * omega * r / vel) / (4. * np.pi * r * M)  
    
    P[index0] = 0.
    
    return P   

def point_source_wave_nonatten(M, omega, r):
    """ green function the nonattenuating media """
    img = 1j
    
    vel = np.sqrt(M)    
       
    P = np.exp(img * omega * r / vel) / (4. * np.pi * r * M)  
    
    return P  
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Nov 27 15:00:48 2019

The point source solution of viscoacoustic wave equation.

@author: sursf
"""

import numpy as np

import matplotlib.pyplot as plt


def ricker_wavelet(fm, t):
    t0 = 1. / fm
    
    tmp = np.pi**2 * fm**2 * (t-t0)**2
    
    ft = (1. - 2. * tmp) * np.exp(-tmp)
    
    tmp0 = np.pi**2 * fm**2 * (0.-t0)**2
    ft0  = (1. - 2. * tmp0) * np.exp(-tmp0)
    
    ft = ft - ft0
    
    return ft



def fourier_time(tdata):
    """ Fourier transform over time. See equation 1 in Hao and Alkhalifah (2019). 
    In fdata_normal, the zero-frequency data is at the center of the array. """
    
    n = len(tdata)
    
    nshift = int(n/2 -1)
    
    fdata = np.fft.ifft(tdata)
    
    fdata_normal = np.roll(fdata, nshift)
       
    return fdata_normal



def invfourier_freq(fdata_normal):
    """ Inverse fourier transform over freq. See equation 2 in Hao and Alkhalifah (2019). 
    In fdata_normal, the zero-frequency data is at the center of the array.
    In tdata, the zero-time data is tdata[0]. """
    
    n = len(fdata_normal)
    
    nshift = int(n/2 - 1)
    
    fdata = np.roll(fdata_normal, -nshift)
    
    tdata = np.fft.fft(fdata)
    
    return tdata
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Nov 27 15:00:48 2019

The point source solution of viscoacoustic wave equation.

@author: sursf
"""

import numpy as np

import matplotlib.pyplot as plt


def ricker_wavelet(fm, t):
    t0 = 1. / fm
    
    tmp = np.pi**2 * fm**2 * (t-t0)**2
    
    ft = (1. - 2. * tmp) * np.exp(-tmp)
    
    tmp0 = np.pi**2 * fm**2 * (0.-t0)**2
    ft0  = (1. - 2. * tmp0) * np.exp(-tmp0)
    
    ft = ft - ft0
    
    return ft



def fourier_time(tdata):
    """ Fourier transform over time. See equation 1 in Hao and Alkhalifah (2019). 
    In fdata_normal, the zero-frequency data is at the center of the array. """
    
    n = len(tdata)
    
    nshift = int(n/2 -1)
    
    fdata = np.fft.ifft(tdata)
    
    fdata_normal = np.roll(fdata, nshift)
       
    return fdata_normal



def invfourier_freq(fdata_normal):
    """ Inverse fourier transform over freq. See equation 2 in Hao and Alkhalifah (2019). 
    In fdata_normal, the zero-frequency data is at the center of the array.
    In tdata, the zero-time data is tdata[0]. """
    
    n = len(fdata_normal)
    
    nshift = int(n/2 - 1)
    
    fdata = np.roll(fdata_normal, -nshift)
    
    tdata = np.fft.fft(fdata)
    
    return tdata
```

### our 1st and 2nd model weighting function coefficients

```python
AlphaBelta =                                  \
np.array([[1.4388052e-01, 1.8931948e-01],     \
          [2.6506214e-02, 2.6022735e-02],     \
          [6.2887118e-03, 5.4548056e-03],     \
          [1.6688598e-03, 1.4214801e-03],     \
          [3.1668719e-04, 4.8742543e-04]])

factor = 0.65 ### fc=40 Hz
# factor = 0.95  ### fc=70 Hz
# factor = 1

### frequency range factor*[1, 50]
tauSigma   = AlphaBelta[:,0].copy()
tauEpsilon = AlphaBelta[:,1].copy() + tauSigma

tauSigma   /= factor
tauEpsilon /= factor

cmodulus_1st2nd_model_init(tauEpsilon, tauSigma)
```

### Medium parameters

```python
### velocity squared
M0 = 3.0 ** 2

### quality factor
Q0 = 100

iQ0 = 1./Q0
```

## Source wavelet

```python
### dominant frequency
fc = 40

#### the number of the sampling points must be even 
n = 4256 * 4

maxt = 2.5 * 2

dt = (maxt - 0.) / (n - 1)


omegamax = np.pi / dt

df = 1. / (n*dt)

t_all = np.arange(0, n) * dt

tdata = ricker_wavelet(fc, t_all)

###### fourier transform ###########
f_all     = (np.arange(0, n) + 1. - n/2.) * df
omega_all = 2.* np.pi * f_all

fdata_normal = fourier_time(tdata)

### remove the alias at zero frequency
### note that after the fourier transform, the data at zero frequency is nonzero and quite large. 
### We need to remove this alias, because the wave amplitude at zero frequency must be zero for a real signal.
index_f0 = int(n/2 -1)
fdata_normal[index_f0] = 0
```

### Wave propagation at distance r = 1 km

```python
### propagation distance
distance = 1

### nonattenuating media
green_freq = point_source_wave_nonatten(M0, omega_all, distance)

wf_freq = np.multiply(fdata_normal, green_freq)


### Kolsky model
green_kolsk_freq = point_source_wave_atten(cmodulus_Kolsk, M0, iQ0, 2.*np.pi*fc, omega_all, distance)

wf_kolsk_freq = np.multiply(fdata_normal, green_kolsk_freq)


### Kjartansson model
green_kjart_freq = point_source_wave_atten(cmodulus_Kjart, M0, iQ0, 2.*np.pi*fc, omega_all, distance)

wf_kjart_freq = np.multiply(fdata_normal, green_kjart_freq)

### our first model
green_1stmodel_freq = point_source_wave_atten(cmodulus_1stmodel, M0, iQ0, 2.*np.pi*fc, omega_all, distance)

wf_1stmodel_freq = np.multiply(fdata_normal, green_1stmodel_freq)


### our second model
green_2ndmodel_freq = point_source_wave_atten(cmodulus_2ndmodel, M0, iQ0, 2.*np.pi*fc, omega_all, distance)

wf_2ndmodel_freq = np.multiply(fdata_normal, green_2ndmodel_freq)



##### inverse fourier transform ###########
wf_t        = invfourier_freq(wf_freq)

wf_kolsk_t = invfourier_freq(wf_kolsk_freq)

wf_kjart_t  = invfourier_freq(wf_kjart_freq)

wf_1stmodel_t  = invfourier_freq(wf_1stmodel_freq)

wf_2ndmodel_t  = invfourier_freq(wf_2ndmodel_freq)
```

```python
##################  Plot the waveform for the nonattenuating model ###############
fig = plt.figure(figsize=(4, 2.5))
ax = fig.add_subplot(1, 1, 1)

fig.subplots_adjust(left=.18, bottom=.16, right=.95, top=.97)

amp_factor = 1

plt.plot(t_all, np.real(amp_factor*wf_t), color='black', lw=2, ls='solid')

plt.grid(True)

plt.xlim(0.2, 0.7)
plt.xlabel("t (s)")
plt.ylabel("P")

plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))

font_size = 12

xlabel = ax.get_xlabel()
ylabel = ax.get_ylabel()

ax.set_xlabel(xlabel, size=font_size)
ax.set_ylabel(ylabel, size=font_size)

ax.tick_params(axis='x', labelsize=font_size)
ax.tick_params(axis='y', labelsize=font_size)


plt.show()

# fname3 = "./figs/wave_withoutQ_fc%s_r%s.pdf" %(str(fc), str(distance))

# fig.savefig(fname3, bbox_inches="tight")


##################  Plot the waveforms for the dissipative models ###############
fig = plt.figure(figsize=(4, 2.5))
ax = fig.add_subplot(1, 1, 1)

fig.subplots_adjust(left=.18, bottom=.16, right=.95, top=.97)


amp_factor = 10**3


plt.plot(t_all, np.real(amp_factor*wf_kjart_t), color='red', lw=2, ls='solid', \
        label="Kjar")

plt.plot(t_all, np.real(amp_factor*wf_kolsk_t), color='lime', lw=2, ls='dashed', \
        label="Kols")

plt.plot(t_all, np.real(amp_factor*wf_1stmodel_t), color='gray', lw=2, ls='dotted', \
        label="1st")

plt.plot(t_all, np.real(amp_factor*wf_2ndmodel_t), color='black', lw=2, ls='dotted', \
        label="2nd")


plt.legend(loc="best", ncol=2)

plt.grid(True)

plt.xlim(0.2, 0.7)
plt.ylim(-3, 6.5)

plt.xlabel("t (s)")
plt.ylabel(r"P ($\times 10^{-3}$)")

plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))

font_size = 12

xlabel = ax.get_xlabel()
ylabel = ax.get_ylabel()

ax.set_xlabel(xlabel, size=font_size)
ax.set_ylabel(ylabel, size=font_size)

ax.tick_params(axis='x', labelsize=font_size)
ax.tick_params(axis='y', labelsize=font_size)

plt.show()

fname4 = "./figs/waves_Qc%s_fc%s_r%s.pdf" %(str(Q0), str(fc), str(distance))

fig.savefig(fname4, bbox_inches="tight")


################### Plot the waveform spectrum #####################
fig = plt.figure(figsize=(4, 2.35))
ax = fig.add_subplot(1, 1, 1)

fig.subplots_adjust(left=.18, bottom=.16, right=.95, top=.97)

amp_factor = 10**5

plt.plot(f_all, np.abs(amp_factor*wf_kjart_freq), color='red', lw=2, ls='solid', \
        label="Kjar")

plt.plot(f_all, np.abs(amp_factor*wf_kolsk_freq), color='lime', lw=2, ls='dashed', \
        label="Kols")

plt.plot(f_all, np.abs(amp_factor*wf_1stmodel_freq), color='gray', lw=2, ls='dotted', \
        label="1st")

plt.plot(f_all, np.abs(amp_factor*wf_2ndmodel_freq), color='black', lw=2, ls='dotted', \
        label="2nd")

plt.legend(loc="best", ncol=1)

plt.grid(True)

plt.xlim(0, 110)
plt.ylim(-0.05, 1.3)

# ylim = ax.get_ylim()
# print("ylim=",ylim)

plt.xlabel("f (Hz)")
plt.ylabel(r"$|\hat{P}|$ ($\times 10^{-5}$)")

plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))

font_size = 12

xlabel = ax.get_xlabel()
ylabel = ax.get_ylabel()

ax.set_xlabel(xlabel, size=font_size)
ax.set_ylabel(ylabel, size=font_size)

ax.tick_params(axis='x', labelsize=font_size)
ax.tick_params(axis='y', labelsize=font_size)

plt.show()

fname5 = "./figs/waves_spectrum_Qc%s_fc%s_r%s.pdf" %(str(Q0), str(fc), str(distance))

fig.savefig(fname5, bbox_inches="tight")
```

```python
##################  Plot the waveform for the nonattenuating model ###############
fig = plt.figure(figsize=(4, 2.5))
ax = fig.add_subplot(1, 1, 1)

fig.subplots_adjust(left=.18, bottom=.16, right=.95, top=.97)

amp_factor = 1

plt.plot(t_all, np.real(amp_factor*wf_t), color='black', lw=2, ls='solid')

plt.grid(True)

plt.xlim(0.9, 1.5)

plt.xlabel("t (s)")
plt.ylabel("P")

plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))

font_size = 12

xlabel = ax.get_xlabel()
ylabel = ax.get_ylabel()

ax.set_xlabel(xlabel, size=font_size)
ax.set_ylabel(ylabel, size=font_size)

ax.tick_params(axis='x', labelsize=font_size)
ax.tick_params(axis='y', labelsize=font_size)


plt.show()

# fname3 = "./figs/wave_withoutQ_fc%s_r%s.pdf" %(str(fc), str(distance))

# fig.savefig(fname3, bbox_inches="tight")


##################  Plot the waveforms for the dissipative models ###############
fig = plt.figure(figsize=(4, 2.5))
ax = fig.add_subplot(1, 1, 1)

fig.subplots_adjust(left=.18, bottom=.16, right=.95, top=.97)


amp_factor = 10**4


plt.plot(t_all, np.real(amp_factor*wf_kjart_t), color='red', lw=2, ls='solid', \
        label="Kjar")

plt.plot(t_all, np.real(amp_factor*wf_kolsk_t), color='lime', lw=2, ls='dashed', \
        label="Kols")

plt.plot(t_all, np.real(amp_factor*wf_1stmodel_t), color='gray', lw=2, ls='dotted', \
        label="1st")

plt.plot(t_all, np.real(amp_factor*wf_2ndmodel_t), color='black', lw=2, ls='dotted', \
        label="2nd")


plt.legend(loc="best", ncol=2)

plt.grid(True)

plt.xlim(0.9, 1.5)
plt.ylim(-5, 8.5)

ax.set_yticks([-4,-2,0,2,4,6,8])

plt.xlabel("t (s)")
plt.ylabel(r"P ($\times 10^{-4}$)")

# plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
# plt.ticklabel_format(axis="y", style="plain",  useOffset=False)

font_size = 12

xlabel = ax.get_xlabel()
ylabel = ax.get_ylabel()

ax.set_xlabel(xlabel, size=font_size)
ax.set_ylabel(ylabel, size=font_size)

ax.tick_params(axis='x', labelsize=font_size)
ax.tick_params(axis='y', labelsize=font_size)

plt.show()

fname4 = "./figs/waves_Qc%s_fc%s_r%s.pdf" %(str(Q0), str(fc), str(distance))

fig.savefig(fname4, bbox_inches="tight")


################### Plot the waveform spectrum #####################
fig = plt.figure(figsize=(4, 2.5))
ax = fig.add_subplot(1, 1, 1)

fig.subplots_adjust(left=.18, bottom=.16, right=.95, top=.97)

amp_factor = 10**6

plt.plot(f_all, np.abs(amp_factor*wf_kjart_freq), color='red', lw=2, ls='solid', \
        label="Kjar")

plt.plot(f_all, np.abs(amp_factor*wf_kolsk_freq), color='lime', lw=2, ls='dashed', \
        label="Kols")

plt.plot(f_all, np.abs(amp_factor*wf_1stmodel_freq), color='gray', lw=2, ls='dotted', \
        label="1st")

plt.plot(f_all, np.abs(amp_factor*wf_2ndmodel_freq), color='black', lw=2, ls='dotted', \
        label="2nd")

plt.legend(loc="best", ncol=2)

plt.grid(True)

plt.xlim(0, 110)
plt.ylim(-0.1, 2.2)

ax.set_yticks([0.0, 0.5, 1.0, 1.5, 2.0])


plt.xlabel("f (Hz)")
plt.ylabel(r"$|\hat{P}|$ ($\times 10^{-6}$)")

plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))

font_size = 12

xlabel = ax.get_xlabel()
ylabel = ax.get_ylabel()

ax.set_xlabel(xlabel, size=font_size)
ax.set_ylabel(ylabel, size=font_size)

ax.tick_params(axis='x', labelsize=font_size)
ax.tick_params(axis='y', labelsize=font_size)

plt.show()

fname5 = "./figs/waves_spectrum_Qc%s_fc%s_r%s.pdf" %(str(Q0), str(fc), str(distance))

fig.savefig(fname5, bbox_inches="tight")
```
